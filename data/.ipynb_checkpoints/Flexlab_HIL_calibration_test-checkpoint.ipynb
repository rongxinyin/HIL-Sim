#### Querying simulation data. Change dates if needed. 

```python
model_cella = pd.read_csv('model_cella.csv', index_col = 0, parse_dates = True)
st = datetime.datetime(2021,8,12,23,45,0)
en = datetime.datetime(2021,8,15,23,45,0)

mask = ((model_cella.index >= st) & (model_cella.index <= en))
model_cella = model_cella[mask]
```

#### gets test data

```python
test_cella = pd.read_csv('test_cella_revised.csv', index_col = 0, parse_dates = True)
test_cellb = pd.read_csv('test_cellb_revised.csv', index_col = 0, parse_dates = True)
```

```python
test_cella['datetime'] = pd.to_datetime(test_cella['datetime'])
test_cellb['datetime'] = pd.to_datetime(test_cellb['datetime'])
```

```python
model_cella = model_cella.interpolate(method = 'linear', limit_direction='both', axis = 0)
test_cella = test_cella.interpolate(method = 'linear', limit_direction='both', axis = 0)

print(len(model_cella.index))
#chwel_df['/x1ab_all/Processed Data/X1/X1-El-Chiller'].interpolate(method = 'linear', limit_direction='forward', axis = 0)
```

### Metrics

```python
test_cella = test_cella.assign(COP='COP') 
test_cella['COP'] = 2.9

test_cellb = test_cellb.assign(COP='COP') 
test_cellb['COP'] = 2.9
```

```python
test_cella = test_cella.assign(chiller = 'Chiller_load') 
test_cella['CellA_Chiller_load'] = (test_cella['1A-Th-Chilled Water']/test_cella['COP']) *-1

test_cellb = test_cellb.assign(chiller = 'Chiller_load') 
test_cellb['Chiller_load'] = (test_cellb['1B-Th-Chilled Water']/test_cellb['COP']) *-1
```

```python
test_cella['1A-SAT-1'] = test_cella['1A-SAT-1'].apply(lambda x: (1.8*x + 32))

model_cella['AHU: Supply Fan Airflow Rate m3/s'] = model_cella['AHU: Supply Fan Airflow Rate'].apply(lambda x: (x/2118.8))
model_cella['AHU: Supply Fan Calc Power'] = model_cella['AHU: Supply Fan Airflow Rate m3/s'].apply(lambda x: (-391*x**2+687*x+140))

```

```python
def calibration_metrics(model, test):
    mse = skl.mean_squared_error(test, model)
    rmse = math.sqrt(mse)
    
    cv_rmse = rmse/(test.mean())
    
    nmbe = np.mean(model.values - test.values) /(test.mean())
    
    mae = skl.mean_absolute_error(test, model)
    cv_test = scipy.stats.variation(test)
    cv_model= scipy.stats.variation(model)
    
    print(nmbe,cv_rmse)
    r2_value = skl.r2_score(test, model)
    
    #return(mse, rmse, cv_rmse, nmbe, mae, cv_test, cv_model, r2_value)
    #return(nmbe,cv_rmse,)
```

```python
AHU_A_power_error = calibration_metrics(model_cella['AHU: Supply Fan Power'], test_cella['1A-El-AHU'])

VAV1_A_dat_error = calibration_metrics(model_cella['VAV: North Zone Discharge Air Temperature'], test_cella['1A-VAV1-DischargeAirTemp'])
VAV2_A_dat_error = calibration_metrics(model_cella['VAV: Core Zone Discharge Air Temperature'], test_cella['1A-VAV2-DischargeAirTemp'])
VAV3_A_dat_error = calibration_metrics(model_cella['VAV: South Zone Discharge Air Temperature'], test_cella['1A-VAV3-DischargeAirTemp'])

VAV_A_airflow_sup_error = calibration_metrics(model_cella['AHU: Supply Fan Airflow Rate'], test_cella['1A-ASAM-1'])
VAV1_A_airflow_sup_error = calibration_metrics(model_cella['VAV: North Zone Airflow Rate'], test_cella['1A-VAV1-DischargeAirflow'])
VAV2_A_airflow_sup_error = calibration_metrics(model_cella['VAV: Core Zone Airflow Rate'], test_cella['1A-VAV2-DischargeAirflow'])
VAV3_A_airflow_sup_error = calibration_metrics(model_cella['VAV: South Zone Airflow Rate'], test_cella['1A-VAV3-DischargeAirflow'])

chw_load_A_error = calibration_metrics(model_cella['AHU: Cooling coil Load'], test_cella['1A-Th-Chilled Water']*-1)

core_temp_A_error = calibration_metrics(model_cella['VAV: Core Zone Room Air Temperature'], test_cella['1A-VAV2-Zn_Temp'])
north_temp_A_error = calibration_metrics(model_cella['VAV: North Zone Room Air Temperature'], test_cella['1A-VAV1-Zn_Temp'])
south_temp_A_error = calibration_metrics(model_cella['VAV: South Zone Room Air Temperature'], test_cella['1A-VAV3-Zn_Temp'])

AHU_A_supply_airtemp_error = calibration_metrics(model_cella['AHU: Supply Air Temperature'], test_cella['1A-SAT-1']*1.8+32)
AHU_A_OADA_error = calibration_metrics(model_cella['AHU: Outdoor Air Damper Position']*100, test_cella['1A-OADA-1-POSN'])
AHU_A_RADA_error = calibration_metrics(model_cella['AHU: Return Air Damper Position']*100, test_cella['1A-RADA-1-POSN'])

```

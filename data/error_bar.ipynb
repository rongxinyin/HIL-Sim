```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('error_data.csv',index_col=0)
print(df)
df=df.drop(index='RA Damper (position)')
```

```python
df.plot.bar(rot=0,figsize=(15,8))
plt.title('Hourly Error Plot',fontsize=20)
plt.xlabel('Variable Name',fontsize=15)
plt.ylabel('Hourly Error (%)',fontsize=15)
plt.legend(fontsize=13)
plt.xticks(rotation='vertical',fontsize=12)
#plt.axhline(-30,color='C1')
plt.axhline(-10,color='C0',linestyle='--')

plt.axhline(30,color='C1',linestyle='--')
plt.axhline(10,color='C0',linestyle='--')


plt.show()

```
